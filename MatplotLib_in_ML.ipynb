```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]

plt.scatter(x,y)
plt.scatter(x2,y2)
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
x=[1,2,3,4,5]
x2=[1,2,3,4,5]
y=[1,2,3,4,5]
y2=[5,4,3,2,1]


plt.plot(x,y,"r-*")
plt.plot(x2,y2,"rp")
plt.show()
```

```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5,6])
y=2*(x**4)+100
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,10,0.1)
y=2*(x**4)+100

plt.plot(x,y)
plt.show()
```

```python
x= np.arange(0,10,0.25)
y=x**4
plt.plot(x,y,color="red",marker="+", linewidth=0.25)
plt.show()
```

```python
x= np.arange(0,10,0.25)
y=x**4
y2=x**3

plt.plot(x,y,color="red", label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.ylabel("X", fontsize=15)
plt.ylabel("Y",fontsize=15)
plt.legend()
plt.show()
```

```python
x= np.arange(0,10,0.25)
y=x**4
y2=x**3

plt.plot(x,y,color="red", label="x^4")
plt.plot(x,y2,color="blue",label="x^3")
plt.ylabel("X", fontsize=15)
plt.ylabel("Y",fontsize=15)
plt.legend()
plt.title("Practice")
plt.axis([0,20,0,10000])
plt.text(14,12500,"Practice")
plt.grid()
plt.show()
```

```python
profit=[10,20,30,25]
labels=["Amazon","Flipkart","Snapdeal","JioMart"]
colors=["red","magenta","pink","blue"]
explode=[0.05,0.1,0,0]

plt.pie(profit, colors=colors,labels=labels, autopct="%.2f%%",explode=explode, counterclock=False, startangle=90)
plt.show()
```
